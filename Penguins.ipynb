```python
# species = ['Adelie', 'Chinstrap', 'Gentoo']
# sex = ['Male', 'Female', nan]
# island = ['Torgersen', 'Biscoe', 'Dream']
peng['sex'] = peng['sex'].fillna(0)
peng['species'].replace(['Adelie', 'Chinstrap', 'Gentoo'], [0,1,2], inplace=True)
peng['island'].replace(['Torgersen', 'Biscoe', 'Dream'], [0,1,2], inplace=True)
peng['sex'].replace(['Male', 'Female'], [1,2], inplace=True)

# print(peng.head())
```

```python
X = peng[['island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']]
y = peng[['species']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,  train_size=0.75, random_state=0, shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(X_test)
acc = clf.score(X_test, y_test)
print(acc)
```

```python
print(classification_report(y_test, y_pred))
```
